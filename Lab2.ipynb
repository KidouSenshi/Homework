```python
# Ввод катетов
a = int(input("Введите длину первого катета (a): "))
b = int(input("Введите длину второго катета (b): "))

# Вычисление гипотенузы по теореме Пифагора
гипотенуза = (a**2 + b**2)**0.5

# Вывод результата
print("Гипотенуза треугольника с катетами", a, "и", b, "равна:", гипотенуза)
```

```python
# Ввод номера урока
номер_урока = int(input("Введите номер урока (от 1 до 10): "))

# Рассчитываем общее время в минутах, учитывая перемены
общее_время_в_минутах = (номер_урока - 1) * 45 + (номер_урока // 2) * 5 + ((номер_урока - 1) // 2) * 15

# Рассчитываем часы и минуты окончания урока
часы = 9 + общее_время_в_минутах // 60
минуты = общее_время_в_минутах % 60

# Вывод результата
print("Урок номер", номер_урока, "заканчивается в", часы, "часов", минуты, "минут.")
```

```python
# Ввод двух чисел
первое_число = int(input("Введите первое число: "))
второе_число = int(input("Введите второе число: "))

# Сравнение чисел и вывод результата
if первое_число > второе_число:
    результат = 1
elif второе_число > первое_число:
    результат = 2
else:
    результат = 0

# Вывод результата
print("Результат:", результат)
```

```python
# Ввод трех целых чисел
первое_число = int(input("Введите первое число: "))
второе_число = int(input("Введите второе число: "))
третье_число = int(input("Введите третье число: "))

# Находим максимальное число с помощью встроенной функции max()
максимальное_число = max(первое_число, второе_число, третье_число)

# Вывод результата
print("Наибольшее число:", максимальное_число)
```

```python
# Ввод трех целых чисел
первое_число = int(input("Введите первое число: "))
второе_число = int(input("Введите второе число: "))
третье_число = int(input("Введите третье число: "))

# Проверка на совпадения чисел
if первое_число == второе_число == третье_число:
    результат = 3  # Все три числа совпадают
elif первое_число == второе_число or второе_число == третье_число or первое_число == третье_число:
    результат = 2  # Два числа совпадают
else:
    результат = 0  # Все числа различны

# Вывод результата
print("Результат:", результат)
```

```python
# Ввод трех чисел
a = int(input("Введите первое число (a): "))
b = int(input("Введите второе число (b): "))
c = int(input("Введите третье число (c): "))

# Сортировка чисел в порядке неубывания
if a > b:
    a, b = b, a
if b > c:
    b, c = c, b
if a > b:
    a, b = b, a

# Вывод упорядоченных чисел
print("Числа в порядке неубывания:", a, b, c)
```

```python
# Ввод коэффициентов квадратного уравнения
a = float(input("Введите коэффициент a: "))
b = float(input("Введите коэффициент b: "))
c = float(input("Введите коэффициент c: "))

# Вычисление дискриминанта
дискриминант = b**2 - 4*a*c

# Проверка наличия корней
if дискриминант > 0:
    корень1 = (-b + дискриминант**0.5) / (2*a)
    корень2 = (-b - дискриминант**0.5) / (2*a)
    print("Уравнение имеет два корня:", корень1, корень2)
elif дискриминант == 0:
    корень = -b / (2*a)
    print("Уравнение имеет один корень:", корень)
else:
    print("Уравнение не имеет действительных корней.")
```
